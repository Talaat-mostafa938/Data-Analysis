## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel(r"Orders.xlsx")
```

```python
#show information about the data
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
#create new columns for day, month and year
df['Day'] = df['Order Date'].dt.day
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
```

```python
#drop the columns that are not needed
df.drop(['Row ID' , 'Order Date'] , axis = 1 , inplace = True)
```

```python
#show information about the data
df.info()
```

```python
#create groupby object
df1 = df.groupby('Region')['Profit'].sum()
df1
```

```python
#show the values of the groupby object
df1.values
```

```python
plt.figure(figsize = (8,8))
sns.barplot( x = df1.index , y = df1.values , palette= 'Blues')
plt.title("Profit by Region")
plt.ylabel("Sum of Profit")
plt.xticks(rotation = 45)
plt.show()
```

```python
# create groupby object
df2 = df.groupby('Month')['Sales'].sum().reset_index()
df2 = df2.astype(int)
```

```python
sns.lineplot(x = df2.iloc[:,0] , y = df2.iloc[:,1] )
plt.xlabel("Monthly Sales")
plt.ylabel("Sum Sales")
plt.xticks(range(1,13) , labels=df2.iloc[:,0])
plt.show()
```

```python
# create groupby object
df3 = df.groupby('Month')['Order ID'].count().reset_index()
```

```python
sns.barplot(x = df3.iloc[:,0] , y = df3.iloc[:,1] , palette='Blues')
plt.title("Monthly Orders")
plt.show()
```

```python
# create groupby object
df4 = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize=(9,9))
sns.barplot(x = df4.iloc[:11,0] , y = df4.iloc[:11,1] , palette='Blues')
plt.title("Top 10 Customer by Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
#create groupby object
df5 = df.groupby('Customer Name')['Profit'].sum().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize=(9,9))
sns.barplot(x = df5.iloc[:11,0] , y = df5.iloc[:11,1] , palette='Blues')
plt.title("Top 10 Customer by Profit")
plt.xticks(rotation=45)
plt.show()
```

```python
# create groupby object
df6 = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize=(9,9))
sns.barplot(y = df6.iloc[:5,0] , x = df6.iloc[:5,1] , palette='Blues')
plt.title("Top 5 Product by Sales")
plt.show()
```

```python
# create groupby object
df7 = df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize=(9,9))
sns.barplot(y = df7.iloc[:5,0] , x = df7.iloc[:5,1] , palette='Blues')
plt.title("Top 5 Product by Profit")
plt.show()
```

```python
# countplot for Ship Mode
sns.countplot(data =df , x = 'Ship Mode' , palette= 'Blues')
plt.show()
```

```python
# countplot for Customer Segment
sns.countplot(data =df , x = 'Customer Segment' , palette= 'Blues')
plt.show()
```

```python
# countplot for Product Category
sns.countplot(data =df , x = 'Product Category' , palette= 'Blues')
plt.show()
```
