```python
# selecting columns with specific data types
# object data types and numeric data types
df_object = df.select_dtypes(include='object')
df_numeric = df.select_dtypes(include='number')
```

```python
df_numeric.shape , df_object.shape
```

```python
for col in df_object.columns:
    print(f"Unique values in {col} : {df_object[col].unique()}")
    print(f"Number of unique values in {col} : {df_object[col].nunique()}")
    print(f"Value counts in {col} : \n{df_object[col].value_counts()}")
    print("-------------------------------------------------------\n")
```

```python
df_numeric.hist(bins=20, figsize=(20, 15), color='blue', edgecolor='black')
plt.suptitle('Histogram of Numeric Features', fontsize=20)
plt.show()
```

```python
for col in df_object.columns:
    plt.figure(figsize=(8, 6))  
    ax = sns.countplot(x=col, data=df_object , palette='mako')
    plt.title(f"Count plot of {col}")
    
    for container in ax.containers:
        ax.bar_label(container, label_type='edge')
    plt.show()
```

```python
pd.crosstab(df["Parental_Involvement"], df["Parental_Education_Level"])
```

```python
ax = pd.crosstab(df["Parental_Involvement"], df["Parental_Education_Level"]).plot(kind="bar", colormap="crest")
for container in ax.containers:
        ax.bar_label(container, label_type='edge')
plt.xticks(rotation=0)
plt.title('Motivation Level vs Hours Studied and Previous Scores')
plt.show()
```

```python
df.groupby('Motivation_Level')[['Hours_Studied','Previous_Scores']].sum()
```

```python
ax = df.groupby('Motivation_Level')[['Hours_Studied','Previous_Scores']].sum().sort_values(by='Previous_Scores', ascending=False).plot(kind='bar', colormap="crest")
for container in ax.containers:
        ax.bar_label(container, label_type='edge')
plt.xticks(rotation=0)
plt.title('Motivation Level vs Hours Studied and Previous Scores')
plt.show()
```
